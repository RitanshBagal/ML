```python
x = np.arange(100)

fig, axis = plt.subplots(2,2)

axis[0,0].plot(x,np.sin(x))
axis[0,0].set_title("Sine Wave")

axis[0,1].plot(x, np.cos(x))
axis[0,1].set_title("Cos Wave")

axis[1,0].plot(x, np.random.random(100))
axis[1,0].set_title("Random Function")

axis[1,1].plot(x, np.log(x))
axis[1,1].set_title("Log Function")

fig.suptitle("Four Plots")
plt.tight_layout()

plt.savefig("fourplots.png")
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # Create a 3D subplot

surf = ax.plot_surface(X, Y, Z, cmap="Spectral")
ax.set_title("3D Plot")
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # Create a 3D subplot

surf = ax.plot_surface(X, Y, Z, cmap="Spectral")
ax.set_title("3D Plot")
ax.view_init(azim=0,elev=90)
```
